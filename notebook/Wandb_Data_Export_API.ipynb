```python
import pandas as pd
import wandb

api = wandb.Api()
entity, project = "akshat_patil", "SLM_vs_LLM_Seq_Clf_Max_length_512"
runs = api.runs(entity + "/" + project,filters={"state": "finished"})

config_list= [] 
for run in runs:
    #  We call ._json_dict to omit large files
    summary_dict = run.summary._json_dict
    summary_list.append(summary_dict)

    # .config contains the hyperparameters.
    # We remove special values that start with _.
    config_dict = {k: v for k, v in run.config.items() if not k.startswith("_")}
    config_dict.update(summary_dict)
    config_list.append(config_dict)
    config_dict["run_id"] = run.id

config_df = pd.DataFrame(config_list)

# Iterate through each run_id and get the max of the GPU columns
for run_id in config_df["run_id"]:
    run = wandb.Api().run(f"{entity}/{project}/{run_id}")
    system_metrics = run.history(stream="events")[['system.gpu.0.memoryAllocated','system.gpu.0.memoryAllocatedBytes','system.gpu.1.memoryAllocated','system.gpu.1.memoryAllocatedBytes']]
    max_gpu_cols = system_metrics.max()
    config_df.loc[config_df["run_id"] == run_id, "max_gpu_0_mem_percent"] = max_gpu_cols["system.gpu.0.memoryAllocated"]
    config_df.loc[config_df["run_id"] == run_id, "max_gpu_0_mem_gb"] = max_gpu_cols["system.gpu.0.memoryAllocatedBytes"]/ (1024 ** 3)
    config_df.loc[config_df["run_id"] == run_id, "max_gpu_1_mem_percent"] = max_gpu_cols["system.gpu.1.memoryAllocated"]
    config_df.loc[config_df["run_id"] == run_id, "max_gpu_1_mem_gb"] = max_gpu_cols["system.gpu.1.memoryAllocatedBytes"]/ (1024 ** 3)

# Create the 'dataset_name' column
config_df['dataset_name'] = config_df['run_name'].apply(lambda x: 
                           'twitter' if 'twitter' in x.lower() 
                           else 'patent' if 'patent' in x.lower()
                           else 'scotus' if 'scotus' in x.lower()
                           else 'ledgar' if 'ledgar' in x.lower()
                           else 'amazon' if 'amazon' in x.lower()
                           else 'unknown')
# Create the 'model_name' column
config_df['model_name'] = config_df['run_name'].apply(lambda x:
                         'google/gemma-7b' if x.str_contains()'google/gemma-7b' in x
                         else 'google/gemma-2b' if 'google/gemma-2b' in x
                         else 'mistralai/Mistral-7B-v0.1' if 'mistralai/Mistral-7B-v0.1' in x
                         else 'meta-llama/Llama-2-7b-hf' if 'meta-llama/Llama-2-7b-hf' in x
                         else 'Qwen/Qwen1.5-1.8B' if 'Qwen/Qwen1.5-1.8B' in x
                         else 'google/flan-t5-base' if 'google/flan-t5-base' in x
                         else 'Qwen/Qwen1.5-7B' if 'Qwen/Qwen1.5-7B' in x
                         else 'google-t5/t5-base' if 'google-t5/t5-base' in x
                         else 'FacebookAI/roberta-base' if 'FacebookAI/roberta-base' in x
                         else 'google/flan-t5-small' if 'google/flan-t5-small' in x
                         else 'google-bert/bert-base-uncased' if 'google-bert/bert-base-uncased' in x
                         else 'google-t5/t5-small' if 'google-t5/t5-small' in x
                         else 'distilbert/distilroberta-base' if 'distilbert/distilroberta-base' in x
                         else 'distilbert/distilbert-base-uncased' if 'distilbert/distilbert-base-uncased' in x
                         else 'microsoft/phi-2' if 'microsoft/phi-2' in x
                         else 'unknown')

# Create the 'finetuning_method' column
config_df['finetuning_method'] = config_df['run_name'].apply(
                                    lambda x:'LoRA' if 'LoRA' in x else 'Full'
                            )

config_df['max_seq_length'] = 512
config_df['gpu_count'] = 2
config_df['gpu'] = 'NVIDIA RTX A6000'
config_df['distributed_training_method'] = config_df['run_name'].apply(lambda x: 
                                          'ddp' if any(model in x.lower() for model in ['bert', 'roberta', 't5', 'flan-t5']) 
                                          else 'deepspeed_stage1')
config_df['lora_rank'] = 128
config_df['lora_alpha'] = 256
config_df['lora_target_modules'] = """'v_proj', 'q_proj', 'down_proj', 'k_proj', 'gate_proj', 'o_proj', 'up_proj'"""

runs_df = config_df[[
    "dataset_name",
    "run_name",
    "model_name",
    "model_type",
    "max_seq_length",
    "finetuning_method",
    "optim",
    "learning_rate",
    "lr_scheduler_type",
    "warmup_steps",
    "warmup_ratio",
    "per_device_train_batch_size",
    "per_device_eval_batch_size",
    "gradient_accumulation_steps",
    "eval_accumulation_steps",
    "num_train_epochs",
    "evaluation_strategy",
    "eval_steps",
    "metric_for_best_model",
    "torch_dtype",
    "gpu",
    "gpu_count",
    "max_gpu_0_mem_percent",
    "max_gpu_0_mem_gb",
    "max_gpu_1_mem_percent",
    "max_gpu_1_mem_gb",
    "distributed_training_method",
    "weight_decay",
    "max_grad_norm",
    "lora_rank",
    "lora_alpha",
    "lora_target_modules",
    "train/loss",
    "eval/loss",
    "test/loss",
    "eval/accuracy",
    "test/accuracy",
    "eval/f1_macro",
    "test/f1_macro",
    "train_runtime",
    "eval/runtime",
    "test/runtime",
    "_runtime"
]]


```

```python

# Create the 'model_name' column
model_names = [
    'google/gemma-7b',
    'google/gemma-2b',
    'mistralai/Mistral-7B-v0.1',
    'meta-llama/Llama-2-7b-hf',
    'Qwen/Qwen1.5-1.8B',
    'google/flan-t5-base',
    'Qwen/Qwen1.5-7B',
    'google-t5/t5-base',
    'FacebookAI/roberta-base',
    'google/flan-t5-small',
    'google-bert/bert-base-uncased',
    'google-t5/t5-small',
    'distilbert/distilroberta-base',
    'distilbert/distilbert-base-uncased',
    'microsoft/phi-2'
]

config_df['model_name'] = config_df['run_name'].apply(lambda x: next((m for m in model_names if m in x), 'unknown'))
```
